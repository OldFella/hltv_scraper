```python
def get_games(url = "https://www.hltv.org/results"):
    browser = webdriver.Firefox()

    print("open website...")
    browser.get("https://www.hltv.org/results")
    print("get data...")
    game_links = browser.find_elements(By.CLASS_NAME, "a-reset")
    match_links = {}
    for e in game_links:
        link = e.get_attribute("href")
        s = link.split(sep = '/')
        if "matches" == s[3]:
            id = s[4]
            match_links[id] = link
    print("close website...")
    browser.close()
    return match_links
```

```python
import re
import pprint

def get_player_stats(data):


    players = data.split('<td class="players">')

    player_dict = {}
    player_count = {}
    
    counts = {1:'total',
              2:'ct',
              3:'t'}

    for player in players:
        start = player.find("/player/")
        if start == -1:
            continue
        end = start + player[start:].find('"')
        _,_,p_id, name = player[start:end].split('/')

        p_id = int(p_id)
        if p_id not in player_count.keys():

            player_dict[p_id] = {'name':name}
            player_count[p_id] = 1

        pattern = r'class="([^"]*((kd|adr|roundSwing|kast|rating)[^"]*(traditional-data|eco-adjusted-data|text-center)[^"]*)[^"]*)">([+-]?[\d\.\-%]+)</td>'
        matches = re.findall(pattern, player)
        stats = {}
        for match in matches:

            classes = match[2]
            value = match[4]
            value = value.replace('%','')

            if 'eco' in match[3]:
                classes = 'e'+classes
                
            if "kd" in classes and "eco" in match[3]:
                k,d = value.split('-')
                stats['ek'] = float(k)
                stats['ed'] = float(d)

            elif "kd" in classes and "eco" not in match[3]:
                k,d = value.split('-')
                stats['k'] = float(k)
                stats['d'] = float(d)
            else:
                stats[classes] = float(value)



        player_dict[p_id][counts[player_count[p_id]]] = stats

        player_count[p_id] += 1

    return player_dict


for score,map in zip(scores,maps):
    pattern = r'id="[^"]*">([^"]*)</'
    match = re.findall(pattern, map)
    stats = get_player_stats(score)
    id = list(match_urls.keys())[0]

    print(f"match_id: {id}, map_name: {match[0]}")
    pprint.pp(stats)
    print()

```
